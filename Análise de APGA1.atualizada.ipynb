## Análise Exploratória de Saúde Neonatal: Indicadores SINASC (Rondônia)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# 1. Configurações Iniciais
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
# 2. Carregamento dos Dados
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

```python
# 3. Análise Exploratória (Foco em Saúde Neonatal)
# Vamos selecionar as colunas de interesse para facilitar a visualização
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

# Criando um painel de visualização profissional
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(16, 18))
axs = axs.flatten()

for i, var in enumerate(variaveis):
    # Usando o gráfico horizontal para variáveis com nomes longos (como GESTACAO)
    order = sinasc[var].value_counts().index
    sns.countplot(
        data=sinasc, 
        y=var, # Mudamos para 'y' para ficar horizontal e elegante
        order=order,
        palette='viridis',
        ax=axs[i]
    )
    axs[i].set_title(f'Distribuição: {var}', fontsize=14, fontweight='bold')
    axs[i].set_xlabel('Frequência')
    axs[i].set_ylabel(None)

# Remove o último eixo que sobrar (caso o número de variáveis seja ímpar)
if len(variaveis) < len(axs):
    fig.delaxes(axs[-1])

plt.tight_layout()
plt.show()
```

```python
# Cruzando APGAR5 com a Escolaridade da Mãe
plt.figure(figsize=(10, 6))
sns.boxplot(data=sinasc, x='APGAR5', y='ESCMAE', palette='coolwarm')
plt.title('Relação entre APGAR5 e Escolaridade da Mãe')
plt.show()
```

### Nesta análise, observamos que a grande maioria dos nascimentos em Rondônia (2019) apresenta índices de APGAR saudáveis. Notamos também uma predominância de gestações que duram entre 37 e 41 semanas...
